### Delete the abnormal data:


- Sudden drops in pressure and abrupt increases in flow are likely due to suction.

### Abnormal Detection

- Go through all the CSV files. If the 'CDGR - PEEP' / 'AVEA - PEEP' / 'SVU - PEEP' is less than or equal to 80% of the PEEP setting, then delete the corresponding rows.

Delete the abnormal data:

- If there are PEEP settings, the abnormal data define as when the actual PEEP values lower than the 80% of the respective PEEP settings.

- If the PEEP settings are absent, then calculate the 1 min (or 5 min) moving average of actual PEEP values, then the abonormal data defines as lower than the 80% of the respective moving average of the actual PEEP values. 

Note: when deleting, delete the corresponding rows include other variables such as PIP, eVT, etc. We use PEEP as a signal to detect abnormal events, such as suction event, etc.

```python
# Add PEEP setting info into df_combined
import os
import pandas as pd

# Define file paths
csv_directory = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/VitalData/01b_PEEP_Settings_294_1st"
df_combined_path = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/df_combined.xlsx"

# Load data from Sheet4 of df_combined
df_sheet4 = pd.read_excel(df_combined_path, sheet_name='Sheet4')

# Loop through all CSV files in the directory
for csv_file in os.listdir(csv_directory):
    if csv_file.endswith('.csv'):
        csv_path = os.path.join(csv_directory, csv_file)

        # Load CSV data
        df = pd.read_csv(csv_path)

        # Define potential PEEP columns
        peep_columns = ["CDGR - Set: PEEP L", "AVEA - Set: PEEP", "SVU - Set: PEEP L"]

        # Find the first non-empty PEEP column
        peep_col = None
        for col in peep_columns:
            if col in df.columns and not df[col].dropna().empty:
                peep_col = col
                break
        # print(peep_col)

        # Proceed only if a valid PEEP column is found
        if peep_col:
            # Extract time and PEEP setting values, and drop rows without data in these columns
            time_col = "Time"  # Assuming the "Time" column exists
            df = df[[time_col, peep_col]].dropna()  # Drop rows without time or PEEP setting
            df[time_col] = pd.to_datetime(df[time_col], errors='coerce')  # Convert to datetime
            df = df.dropna(subset=[time_col])  # Remove rows with invalid times
            # print(df.shape)

            # Check if df has rows to avoid index errors
            if not df.empty:
                # Initialize lists to hold each period of PEEP settings
                peep_periods = []
                start_time = df[time_col].iloc[0]
                current_peep = df[peep_col].iloc[0]

                for i in range(1, len(df)):
                    if df[peep_col].iloc[i] != current_peep:
                        # Record the current period
                        stop_time = df[time_col].iloc[i - 1]
                        peep_periods.append((start_time, stop_time, current_peep))
                        
                        # Update start time and current PEEP setting
                        start_time = df[time_col].iloc[i]
                        current_peep = df[peep_col].iloc[i]

                # Append the last period
                peep_periods.append((start_time, df[time_col].iloc[-1], current_peep))
                
                # Find the row in df_sheet4 corresponding to the current CSV file's filename
                file_identifier = csv_file
                match_row = df_sheet4[df_sheet4["FileName_PEEP_Settings_1st"] == file_identifier]

                # If there's a match, insert the data into new columns in df_sheet4
                if not match_row.empty:
                    row_index = match_row.index[0]
                    for i, (start, stop, peep) in enumerate(peep_periods, start=1):
                        # Add columns dynamically
                        df_sheet4.at[row_index, f"PEEP_Settings_1st_Time_Start_{i}"] = start
                        df_sheet4.at[row_index, f"PEEP_Settings_1st_Time_Stop_{i}"] = stop
                        df_sheet4.at[row_index, f"PEEP_Settings_1st_{i}"] = peep

# Save the modified DataFrame as Sheet5
with pd.ExcelWriter(df_combined_path, engine="openpyxl", mode="a") as writer:
    df_sheet4.to_excel(writer, sheet_name="Sheet5", index=False)
```

```python
# Calculate the 5-min moving average when PEEP Settings are missing
import os
import pandas as pd

# Define paths
csv_directory = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/VitalData/01_Vital_Raw_282_1st/OSI_285_1st/TWs_284_1st/PEEP_Settings_284_1st/"
destination_directory = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/VitalData/01_Vital_Raw_282_1st/OSI_285_1st/TWs_284_1st/PEEP_Settings_284_1st_v2/"

# Ensure destination directory exists
os.makedirs(destination_directory, exist_ok=True)

# Define PEEP columns to look for
peep_columns = ["CDGR - PEEP", "AVEA - PEEP", "SVU - PEEP"]

# Loop through all CSV files in the source directory
for csv_file in os.listdir(csv_directory):
    if csv_file.endswith('.csv'):
        csv_path = os.path.join(csv_directory, csv_file)
        df_csv = pd.read_csv(csv_path)

        # Identify the actual PEEP column dynamically
        actual_peep_column = next((col for col in peep_columns if col in df_csv.columns), None)
        if not actual_peep_column:
            print(f"Warning: No PEEP column found in {csv_file}. Skipping file.")
            continue

        # Ensure Time column is properly parsed
        if "Time" in df_csv.columns:
            df_csv["Time"] = pd.to_datetime(df_csv["Time"], errors="coerce")
        else:
            print(f"Warning: Time column not found in {csv_file}. Skipping file.")
            continue

        # Set Time as the index for rolling operations
        df_csv = df_csv.set_index("Time")

        # Check if PEEP_Settings is empty
        if df_csv["PEEP_Settings"].isnull().all():
            print(f"PEEP_Settings is empty in {csv_file}. Calculating 5-minute moving average...")
            
            # Calculate the 5-minute moving average and assign to a new column PEEP_MovingAvg
            df_csv["PEEP_MovingAvg"] = df_csv[actual_peep_column].rolling("5T").mean()
        else:
            # If PEEP_Settings is not empty, ensure PEEP_MovingAvg is filled with NaN
            df_csv["PEEP_MovingAvg"] = pd.NA

        # Reset the index back to a regular column
        df_csv = df_csv.reset_index()

        # Save the modified CSV to the destination directory
        destination_path = os.path.join(destination_directory, csv_file)
        df_csv.to_csv(destination_path, index=False)
        # print(f"Processed and saved {csv_file} to the destination directory.")

print("Processing complete. All files have been checked and modified if needed.")
```

```python
# Abnormal data deletion
import os
import pandas as pd
import numpy as np

# Define paths
csv_directory = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/VitalData/01_Vital_Raw_282_1st/OSI_285_1st/TWs_284_1st/PEEP_Settings_284_1st/"
destination_directory = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/VitalData/01_Vital_Raw_282_1st/OSI_285_1st/TWs_284_1st/Abnormal_Detection_282_1st/"

# Ensure destination directory exists
os.makedirs(destination_directory, exist_ok=True)

# Define PEEP columns to look for
peep_columns = ["CDGR - PEEP", "AVEA - PEEP", "SVU - PEEP"]

# Loop through all CSV files in the source directory
for csv_file in os.listdir(csv_directory):
    if csv_file.endswith('.csv'):
        csv_path = os.path.join(csv_directory, csv_file)
        df_csv = pd.read_csv(csv_path)

        # Identify the actual PEEP column dynamically
        actual_peep_column = next((col for col in peep_columns if col in df_csv.columns), None)
        # print(actual_peep_column)
        if not actual_peep_column:
            print(f"Warning: No PEEP column found in {csv_file}. Skipping file.")
            continue

        # Ensure Time column is properly parsed
        if "Time" in df_csv.columns:
            df_csv["Time"] = pd.to_datetime(df_csv["Time"], errors="coerce")
        else:
            print(f"Warning: Time column not found in {csv_file}. Skipping file.")
            continue

        # Set Time as the index for rolling operations
        df_csv = df_csv.set_index("Time")

        # Check if PEEP_Settings is empty
        if df_csv["PEEP_Settings"].isnull().all():
            ##########################################################################
            # Apply a 1-min / 5-min moving average if PEEP_Settings is empty
            df_csv["PEEP_MovingAvg"] = df_csv[actual_peep_column].rolling("5T").mean()
            ##########################################################################

            # Calculate 80% of the moving average
            df_csv["PEEP_Threshold"] = df_csv["PEEP_MovingAvg"] * 0.8

            # Delete rows where actual PEEP values are less than 80% of the moving average
            df_csv = df_csv[~(df_csv[actual_peep_column] < df_csv["PEEP_Threshold"])]
        else:
            # Calculate 80% of PEEP_Settings if it has data
            df_csv["PEEP_Threshold"] = df_csv["PEEP_Settings"] * 0.8

            # Delete rows where actual PEEP values are less than 80% of PEEP_Settings
            df_csv = df_csv[~(df_csv[actual_peep_column] < df_csv["PEEP_Threshold"])]

        # Reset the index back to a regular column
        df_csv = df_csv.reset_index()

        # Save the modified CSV to the destination directory
        destination_path = os.path.join(destination_directory, csv_file)
        df_csv.to_csv(destination_path, index=False)
```

```python
import os
import pandas as pd

# Paths to the directories
vital_path = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/VitalData/01_Vital_Raw_282_1st/OSI_285_1st/TWs_284_1st/PEEP_Settings_284_1st/Abnormal_Detection_282_1st"
waveform_path = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/WaveformData/01_Waveform_Raw_248_1st/TWs_243_1st"
output_path = os.path.join(waveform_path, "Abnormal_Detection_243_1st")

# Create output directory if it doesn't exist
os.makedirs(output_path, exist_ok=True)

# Helper function to extract common identifier from file name
def extract_identifier(filename):
    return "_".join(filename.split("_")[:3]) + ".csv"

# Get list of files from both directories
vital_files = os.listdir(vital_path)
waveform_files = os.listdir(waveform_path)

# Match files by identifier
vital_dict = {extract_identifier(f): f for f in vital_files}
waveform_dict = {extract_identifier(f): f for f in waveform_files}

matched_files = [(vital_dict[key], waveform_dict[key]) for key in vital_dict if key in waveform_dict]

# Process each matched pair
for vital_file, waveform_file in matched_files:
    try:
        # print(f"Processing files: {vital_file}, {waveform_file}")
        # Load the CSV files
        vital_df = pd.read_csv(os.path.join(vital_path, vital_file))
        waveform_df = pd.read_csv(os.path.join(waveform_path, waveform_file))

        # Ensure 'Time' column exists
        if 'Time' not in vital_df.columns or 'Time' not in waveform_df.columns:
            print(f"Missing 'Time' column in {vital_file} or {waveform_file}. Skipping.")
            continue

        # Convert 'Time' columns to datetime
        vital_df['Time'] = pd.to_datetime(vital_df['Time'])
        waveform_df['Time'] = pd.to_datetime(waveform_df['Time'])

        # Extract seconds from 'Time'
        vital_seconds = vital_df['Time'].dt.floor('S').unique()  # Unique seconds in vital file
        waveform_seconds = waveform_df['Time'].dt.floor('S')    # Seconds in waveform file

        # Identify rows to delete
        rows_to_delete = waveform_df[waveform_seconds.isin(vital_seconds)].index

        # Remove rows from the Waveform DataFrame
        modified_waveform_df = waveform_df.drop(index=rows_to_delete)

        # Save the modified waveform file
        modified_waveform_file = os.path.join(output_path, waveform_file)
        modified_waveform_df.to_csv(modified_waveform_file, index=False)

        # print(f"Processed and saved: {waveform_file}")
    except Exception as e:
        print(f"Error processing files {vital_file} and {waveform_file}: {e}")

print("Processing completed.")
```
