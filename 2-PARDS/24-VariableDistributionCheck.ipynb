### Check the variable distributions and ensure that the variables are not skewed when using the mean to represent them.

Check 1st hour OSI values

```python
import pandas as pd
import os
from scipy.stats import skew

# Path to the directory containing the CSV files
csv_directory = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/VitalData/01_Vital_Raw_282_1st/OSI_285_1st/TWs_284_1st/PEEP_Settings_284_1st/Abnormal_Detection_282_1st"

# Threshold for skewness; a common threshold is ±0.5 (low skewness)
skew_threshold = 0.5

# Loop through all CSV files in the directory
results = []
for file_name in os.listdir(csv_directory):
    if file_name.endswith(".csv"):  # Only process CSV files
        file_path = os.path.join(csv_directory, file_name)
        
        # Read the CSV file
        df_csv = pd.read_csv(file_path)
        
        # Check if "OSI" column exists
        if "OSI" in df_csv.columns:
            # Calculate skewness of the "OSI" column
            osi_skewness = skew(df_csv["OSI"].dropna())  # Drop NaN values for skewness calculation
            
            # Determine if skewness is within acceptable limits
            skewed = abs(osi_skewness) > skew_threshold
            status = "Skewed" if skewed else "Not Skewed"
            
            # Append results
            results.append({
                "File Name": file_name,
                "Skewness": osi_skewness,
                "Status": status
            })
        else:
            print(f"Column 'OSI' not found in file: {file_name}")

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Filter rows with skewness out of range
skewed_rows = results_df[results_df["Status"] == "Skewed"]

# Print rows where skewness is out of range
if not skewed_rows.empty:
    print("Files with skewed OSI distributions:")
    print(skewed_rows)
else:
    print("No files with skewed OSI distributions found.")

# # Save all results to a CSV file for review
# output_path = os.path.join(csv_directory, "OSI_Distribution_Check.csv")
# results_df.to_csv(output_path, index=False)

print("Analysis complete. Results saved to 'OSI_Distribution_Check.csv'.")
```
