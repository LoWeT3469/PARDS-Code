**Oxygenation Saturation Index (OSI)** $= \frac{Paw \space \times \space FiO_{2}}{SpO_{2}}$

```python
# Vital_Raw_292_1st / Vital_Raw_226_12th
import pandas as pd
import numpy as np
import os

# Define the directory path
# directory = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/VitalData/01_Vital_Raw_292_1st/"
directory = "/nfs/turbo/med-kayvan-lab/Projects/PARDS/02-Data/TRAPPARDS/VitalData/02_Vital_Raw_226_12th/"
output_directory = os.path.join(directory, "OSI")

# Create output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Loop through each CSV file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        try:
            # Read the CSV file into a DataFrame
            filepath = os.path.join(directory, filename)
            df = pd.read_csv(filepath)
            
            # Check which ventilator type columns are non-empty
            if not df[['CDGR - Paw', 'CDGR - FiO₂']].isnull().all().any():
                # Calculate OSI for CDGR
                df["OSI"] = df["CDGR - Paw"] * df["CDGR - FiO₂"] / df["GE - SpO2 1"]
            elif not df[['AVEA - Paw', 'AVEA - FiO₂']].isnull().all().any():
                # Calculate OSI for AVEAA
                df["OSI"] = df["AVEA - Paw"] * df["AVEA - FiO₂"] / df["GE - SpO2 1"]
            elif not df[['SVU - Paw', 'SVU - FiO₂']].isnull().all().any():
                # Calculate OSI for SVU
                df["OSI"] = df["SVU - Paw"] * df["SVU - FiO₂"] / df["GE - SpO2 1"]
            else:
                print(f"No valid ventilator data found in {filename}. Skipping...")
                continue
            
            # Drop inf values from OSI column
            df = df.replace([np.inf, -np.inf], np.nan)  # Replace inf with NaN
            df = df.dropna(subset=["OSI"], how="any")  # Drop rows with NaN in OSI column
            
            # # Calculate OSI_1st_avg column
            # df["OSI_1st_avg"] = df["OSI"].mean()

            # Calculate OSI_12th_avg column
            df["OSI_12th_avg"] = df["OSI"].mean()
            
            # Remove empty columns
            df = df.dropna(axis=1, how="all")
            
            # Select only the numeric columns
            numeric_cols = df.select_dtypes(include=[np.number])

            # Count inf values in each column
            inf_counts = numeric_cols.apply(lambda x: np.isinf(x).sum())

            # print(f"Number of inf values in each column for {filename}:")
            # print(inf_counts)
            
            # Save the modified DataFrame to the new CSV file in the "OSI" folder
            output_filepath = os.path.join(output_directory, filename)
            df.to_csv(output_filepath, index=False)
        
        except Exception as e:
            # Print the error message and the filename that caused the error
            print(f"Error processing file {filename}: {e}")

print("Files modified and saved to 'OSI' folder successfully.")
```
